Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.

```python
cake_recipes = pd.read_csv('untidy-data/cake_recipes.csv')
```

```python
cake_recipes.head()
```

```python
cake_recipes.info()
```

Tidy the data as necessary.

```python
cake_recipes.shape
```

```python
cake_recipes = cake_recipes.rename(columns={"recipe:position": "recipe"})
```

```python
cake_recipes
```

```python
new_cake_recipes = cake_recipes.recipe.str.split(':', expand=True)
new_cake_recipes.columns = ['Recipes', 'Position']
new_cake_recipes = pd.concat([cake_recipes, new_cake_recipes], axis=1).drop(columns='recipe')
```

```python
new_cake_recipes
```

```python
# df.melt(id_vars=['a', 'b'], value_vars=['x', 'y'], var_name='original_column_name', value_name='original_value')
# new_cake_recipes.melt(id_vars=['Recipes'])
```

```python
cake_recipes = new_cake_recipes.melt(id_vars=['Recipes', 'Position'])
```

```python
cake_recipes
```

```python
cake_recipes = cake_recipes.rename(columns={"value": "taste_score"})
```

```python
cake_recipes = cake_recipes.rename(columns={"variable": "temp"})
```

```python
cake_recipes.head()
```

```python
cake_recipes.groupby('Recipes').taste_score.mean()
```

Which oven temperature, on average, produces the best results? 275



```python
cake_recipes.groupby('temp').taste_score.mean()
```
