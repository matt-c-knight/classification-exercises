Attendance Data

Load the attendance.csv file and calculate an attendnace percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
attendance = pd.read_csv('attendance.csv')
```

```python
attendance
```

```python
attendance.info()
```

```python
attendance = attendance.rename(columns={'Unnamed: 0': 'student_name'})
```

```python
attendance
```

```python
attendance = attendance.melt(id_vars='student_name')
```

```python
attendance
```

```python
attendance = attendance.drop(columns='variable')
```

```python
attendance
```

```python
attendance['value'] = attendance['value'].replace(['P','A','T','H'], [1,0,.9,.5])
```

```python
attendance
```

```python
attendance = attendance.groupby('student_name').mean()
```
